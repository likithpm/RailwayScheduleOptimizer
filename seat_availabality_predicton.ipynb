```python
# splitting the data into x an y values
x=data.drop("labels",axis=1)
y=data["labels"]
```

```python
# to get max of every column
columns=data.columns

for col in columns:
  print(f"{col} max:{data[col].max}")
```

```python
# giving more weight to the first 3 columns
data["travelClass"]=data["travelClass"]*3
data["bookingStatus"]=data["bookingStatus"]*3
data["status1Day"]=data["status1Day"]*3
```

```python
#split into train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=42)
```

```python
data.head()
```

```python
#importing  metrics
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

#importing the libraries
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
```

```python
#choosing the right model dictionary
models={
    "Logistic Regression":LogisticRegression(max_iter=1000),
    "Random Forest":RandomForestClassifier(),
    "Gradient Boosting":GradientBoostingClassifier(),
    "Support Vector Classifier":SVC(),
    "K-Nearest Neighbors":KNeighborsClassifier(),
    "Decision Tree":DecisionTreeClassifier(),
    "Naive Bayes":GaussianNB(),
    "XGBoost":xgb.XGBClassifier(use_label_encoder=False,eval_metric="mlogloss")
}
```

```python
#Loop through the model
for model_name,model in models.items():
  print(f"Training {model_name}")

  #training the model
  model.fit(x_train,y_train)

  #make prediction
  y_preds=model.predict(x_test)

  #evaluation metrics
  accuracy=accuracy_score(y_test,y_preds)
  precision=precision_score(y_test,y_preds)
  recall=recall_score(y_test,y_preds)
  f1=f1_score(y_test,y_preds)
  conf_matrix=confusion_matrix(y_test,y_preds)

  #print the metrics results
  print(f"Accuracy:{accuracy:.4f}")
  print(f"Precision:{precision:.4f}")
  print(f"Recall:{recall:.4f}")
  print(f"F1:{f1:.4f}")
  print(f"Confussion Matris:\n{conf_matrix}\n")
```

```python
# Custom loss function (logistic regression loss with feature importance weighting)
def weighted_logistic_loss(y_true, y_pred):
    weights = np.array([3, 3, 3, 1, 1, 1])  # More importance to first 3 features
    # Logistic loss function
    preds = 1.0 / (1.0 + np.exp(-y_pred))

    # Calculate gradient and hessian
    grad = preds - y_true
    hess = preds * (1.0 - preds)

    # Apply weights (you may want to modify how these weights are applied based on features)
    weighted_grad = grad * np.mean(weights)  # Applying the weight (averaging for simplicity)
    weighted_hess = hess * np.mean(weights)  # Apply the weight here too

    return weighted_grad, weighted_hess

# Train an XGBoost model using the custom loss function
model = xgb.XGBClassifier(objective=weighted_logistic_loss)
model.fit(x_train, y_train)
```
